## import nesessery libebery

```python
import numpy as np
from tensorflow import keras
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

## load data

```python
iris_data = load_iris() # load the iris dataset
x = iris_data.data
target= iris_data.target.reshape(-1, 1) # Convert data to a single column
print(x.shape)
print(target.shape)
```

## regulization

```python
y=keras.utils.to_categorical(target)
```

## Split the data for training and testing

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

## Build the model

```python
model=Sequential()
# Add an initial layer with 4 input nodes, and a hidden layer with 16 nodes.
model.add(Dense(16,activation='relu',input_shape=(4,))),
model.add(Dense(16,activation='relu')),
model.add(Dropout(0.5)),
model.add(Dense(3,activation='softmax'))
```

## Adam optimizer

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

##Train

```python
history = model.fit(train_x,train_y,
          epochs=200,
          batch_size=5, validation_split=0.2)
```

## model evaluate

```python
# Test on unseen data
results = model.evaluate(test_x, test_y)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```
