```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("youtube.csv",encoding='latin1')
print(df.to_string())
```

```python
gb=df.groupby("Country").count()
gb=gb["subscribers"]
gb=gb.sort_values(ascending=False)
gb=gb.head(10)
gb.plot(kind="pie",y="subscribers")
plt.title("total subscribers based on country")
```

```python
plt.figure(figsize=(15,7))
df=pd.read_csv("youtube.csv",encoding='latin1')
gb=df.groupby("Country").count()
gb=gb["subscribers"]
gb=gb.sort_values(ascending=False)
gb=gb.head(10)
gb.plot(kind="bar")
plt.ylabel("subscribers")
plt.xlabel("country")
plt.title("total subscribers based on country")
plt.show()
```

```python
group_by2=df.groupby("category")["video views"].sum()
group_by2.head(5).plot(kind="pie")
plt.title("total views based on category")
plt.show()
```

```python
top10category=df["category"].value_counts()
top10category.head(10).plot(kind="pie")
plt.title("Top 10 Category in youTube")
plt.ylabel(" ")
plt.show()
```
